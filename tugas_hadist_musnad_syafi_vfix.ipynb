```python
import pandas as pd
import re
import spacy
from deep_translator import GoogleTranslator
```

```python
musnad_syafii = pd.read_csv("D:/RoadMap Kuliah/Semester 7/Quran Hadis Mining/tugasku/dataset hadist/musnad_syafii.csv",sep=";")
```

```python
def translate_to_english(text):
    translated_text = GoogleTranslator(source='indonesian', target='english').translate(text)
    return translated_text

def translate_to_indonesian(text):
    translated_text = GoogleTranslator(source='english', target='indonesian').translate(text)
    return translated_text

def find_name_with_spacy(paragraph):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(paragraph)
    names = [entity.text for entity in doc.ents if entity.label_ == 'PERSON']
    return names
```

```python
for i in range(len(musnad_syafii)):
    musnad_syafii.loc[i, "terjemah"] = musnad_syafii.loc[i, "terjemah"].replace(f"Musnad Syafi'i {i+1}: ", "")
    musnad_syafii.loc[i, "terjemah"] = re.sub(r'\b\d+\b', '', musnad_syafii.loc[i, "terjemah"])
```

```python
musnad_syafii['terjemah_english'] = musnad_syafii['terjemah'].apply(translate_to_english)
```

```python
hasil_musnad_syafii = musnad_syafii[["id","kitab"]].copy()
```

```python
hasil_musnad_syafii
```

```python
hasil_perawi = []
hasil_matan_inggris = []

kata_kunci = [
    '\r\n\r\n', 'say :', ': That', 'said,', 'stated:', 'say:', 'said:',
    '. That', ', ""', ';', ', "', 'bersabda', 'about',
    "sallallaahu 'alaihi wa sallam", "he said", "Rasulullah",
    "from the Prophet", "with a similar editorial", "with the same",
    "with similar", "namely a similar hadith", "with a hadith similar",
    "with an editorial similar","Umar bin Al Khaththab wrote a letter",":", "such as", "that", "regarding"
]

for i in range(len(musnad_syafii)):
    matan = ""
    terjemah = musnad_syafii_v2.loc[i, "terjemah_english"]
    perawi = "program gagal"
    for kunci in kata_kunci:
        if kunci in terjemah:
            kalimat_terpisah = terjemah.split(kunci)
            perawi = find_name_with_spacy(kalimat_terpisah[0])
            matan = kunci.join(kalimat_terpisah[1:])
            break
    
    if matan == "":
        matan = "program gagal"
        
    hasil_perawi.append(perawi)
    hasil_matan_inggris.append(matan)
```

```python
hasil_musnad_syafii["hasil_perawi"] = hasil_perawi
hasil_musnad_syafii["matan_inggris"] = hasil_matan_inggris
hasil_musnad_syafii["terjemah"] = musnad_syafii_v2["terjemah"]
```

```python
jumlah_kemunculan = hasil_musnad_syafii["hasil_perawi"].apply(lambda x: x.count("program gagal")).sum()
print(f"Jumlah program error: {jumlah_kemunculan}")
```

```python
hasil_musnad_syafii["matan_indo"] = hasil_musnad_syafii["matan_inggris"].apply(translate_to_indonesian)
```

```python
hasil_musnad_syafii
```

```python
def clean_list(lst, word):
    return list(filter(lambda x: word.lower() not in x.lower(), lst))

# Membersihkan nilai-nilai dari kata "Rasulullah" atau "Prophet"
for idx, perawi_list in enumerate(hasil_musnad_syafii["hasil_perawi"]):
    hasil_musnad_syafii["hasil_perawi"][idx] = clean_list(perawi_list, 'rasulullah')
    hasil_musnad_syafii["hasil_perawi"][idx] = clean_list(hasil_musnad_syafii["hasil_perawi"][idx], 'prophet')
```
